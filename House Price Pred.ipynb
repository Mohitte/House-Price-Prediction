```python
# Max Depth
```

```python
max_depth=[]
acc_list=[]
for i in range(1,30):
    dt1=DecisionTreeRegressor(max_depth=i)
    dt1.fit(xtrain,ytrain)
    ypred=dt1.predict(xtest)
    acc=r2_score(ytest,ypred)
    max_depth.append(i)
    acc_list.append(acc)
    print(f'max depth={i}, accuracy = {acc*100}')
```

```python
max_acc=acc_list[0]
for i in range(1,len(acc_list)):
    if (max_acc < acc_list[i]):
        max_acc=acc_list[i]
max_index=acc_list.index(max_acc)
print(f'max acc index = {max_index+1}')
```

```python
dt2=DecisionTreeRegressor(max_depth=8)
mymodel(dt2)
```

```python
dt2.score(xtrain,ytrain)*100
```

```python
#Min sample Split
```

```python
min_sample=[]
acc_list=[]
for i in range(2,50):
    dt3=DecisionTreeRegressor(min_samples_split=i)
    dt3.fit(xtrain,ytrain)
    ypred=dt3.predict(xtest)
    acc1=r2_score(ytest,ypred)
    min_sample.append(i)
    acc_list.append(acc1)
    print(f'Min saple split= {i}, acc = {acc1*100}')
```

```python
min_acc=acc_list[0]
for i in range(1, len(acc_list)):
    if (acc_list[i]>min_acc):
        min_acc=acc_list[i]
min_acc_index=acc_list.index(min_acc)
print(f'min acc index = {min_acc_index+2}')
```
